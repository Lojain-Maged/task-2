```python
email = "Amit_ml@gmail.edu"

if email.count("@") != 1 or "." not in email.split("@")[1]:
    print("Invalid email")
else:
    username = email.split("@")[0]
    domain = email.split("@")[1].rsplit(".", 1)[0]

    domain_type = "Other Domain"
    if email.endswith(".com"):
        domain_type = "Commercial Domain"
    if email.endswith(".edu"):
        domain_type = "Educational Domain"

    print("Username:", username)
    print("Domain:", domain)
    print("Domain Type:", domain_type)

```

```python
msg = "&&&**$gnirtS PLIO!!@1234"

c = msg.split("$")[1].split("!!")[0].strip()

words = c.split()
words[0] = words[0][::-1]
words[1] = words[1].replace("I", "E").replace("O", "U")

print(words[0] + " " + words[1])
```
